## Clean DataFrame

```python
#clean 'main_df'
#df.loc[(df["B"] > 50) & (df["C"] == 900), "A"]
clean_main_df = main_df.loc[(main_df["Country"] == 'China') | (main_df["Country"] == 'Mainland China')]
clean_main_df
```

```python
main_df['Country'] = main_df['Country'].replace(
    {'Mainland China': 'China',
    'Hong Kong': 'China',
    'Macau': 'China',
    'Taiwan':'China'})

main_df
```

```python
#clean confirmed_df; only show 'Province/State', 'Country/Region', 'Lat', 'Long'; then make copy with 'copy()'
clean_confirmed_df = confirmed_df[["Province/State", "Country/Region", "Lat", "Long"]].copy()
clean_confirmed_df
```

```python
#rename Province/State 'NaN' entries with 'Unknown'
# filled_cleaned_confirmed_df = new_confirmed_df.replace(np.NaN, "Unknown", regex=True)
# values = {'Province/State': "Unknown"} 
# # new_confirmed_df.fillna(value=values)
```

```python
#clean deaths_df; replace 'Mainland China', to 'China'; 'Hong Kong' to China, "Macau" to China, "Taiwan" to China
deaths_df['Country/Region'] = deaths_df['Country/Region'].replace(
    {'Mainland China': 'China',
    'Hong Kong': 'China',
    'Macau': 'China'})
deaths_df = deaths_df.rename(columns={"2/20/20": "Deaths"})
deaths_df
```

```python
deaths_df['Province/State'] = deaths_df['Province/State'].replace(
    {'Cruise Ship': 'Diamond Princess cruise ship',
     'Chicago': 'Chicago, IL',
    'California':'Sacramento, CA',
    'San Diego County, CA':'San Diego, CA',
    'Washington': 'Seattle, WA',
    'Illinois': 'Chicago, IL',
     'Ontario': 'Toronto, ON',
     'Victoria':'Victoria, BC',
     'Arizona': 'Tempe, AZ'
    })
deaths_df
```

```python
deaths_df.to_csv("output/cleaned_deaths.csv")
```
